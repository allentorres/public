```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-3,3,.001) # arange(x1,x2,d) shows the list of values from x1 to x2 with a difference of d
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,.5,.5)) #can plot two graphs at once
plt.show()
plt.savefig('C:\\MLCourse\\hello.png',format='png')
```

adjusting the axes

```python
axes = plt.axes()
axes.set_xlim([-5,5])
axes.set_ylim([0,1])
axes.set_xticks([-5,-3,-1,0,1,3,5])
axes.set_yticks([0,.2,.4,.6,.8,1])
axes.grid()
x=np.arange(-5,5,.01)
plt.xlabel('Crombs')
plt.ylabel('Probability')
plt.plot(x,norm.pdf(x),'b-') #blue line
plt.plot(x,norm.pdf(x,.5,.5),'r:') #red dots
plt.plot(x,norm.pdf(x,-.5,2),'g--') #green dashes
plt.legend(['Rats','Snitches','Pigs'],loc=4)
plt.show()
```

Pie chart

```python
values = [12,2,24,25,23] #dependent values
colors = ['r','g','b','c','m'] #+ cyan and magenta
explode = [0,.2,0,0,0] #how much each is highlighted
labels = ['India','Russia','USA','UK','Thailand'] #independent variables
plt.pie(values,colors=colors,labels=labels,explode=explode)
plt.title('Student Locations')
plt.show()
```

Bar chart

```python
values = [12,2,24,25,23] #dependent values
colors = ['r','g','b','c','m'] #+ cyan and magenta
#plt.bar(range(0,5),values,color=colors)
plt.bar(['India','Russia','USA','UK','Thailand'],values,color=colors)
plt.show()
```
