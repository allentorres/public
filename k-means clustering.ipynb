K-Means clustering Example
===

We create fake data that includes people clustered by income and age

```python
import random
import numpy as np
def createClusteredData(N,k):
    random.seed(100)
    pointsPerCluster = float(N)/k
    X=[]
    for i in range(k):
        incomeCentroid = random.uniform(20000,200000)
        ageCentroid = random.uniform(20,70)
        for j in range(int(pointsPerCluster)):
            X.append([np.random.normal(incomeCentroid,10000),np.random.normal(ageCentroid,2)])
    X = np.array(X)
    return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

data = createClusteredData(100,5)

model = KMeans(n_clusters=5)

# scaling data to normalize
model = model.fit(scale(data))

print(model.inertia_)
plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1], c = model.labels_.astype(np.float))
plt.show()
```

```python
for i in range(1,10):
    model = KMeans(n_clusters=i)
    model = model.fit(scale(data))
    print(i, ' ', model.inertia_)
```
