```python
import pandas as pd
df=pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")
gear_counts=df['# Gears'].value_counts() #creates a dictionary for a histogram
gear_counts.plot(kind='bar') #creates a bar chart
```

using Seaborn. Just makes matplotlib prettier

```python
import seaborn as sns
sns.set()
```

```python
gear_counts.plot(kind='bar') #creates a bar chart but now uses seaborn
```

```python
df.head() #gets first 5 rows
```

```python
#distplot
sns.distplot(df['CombMPG']) #creating a smooth distribution with the histogram
```

```python
#pairplot
df2=df[['Cylinders','CityMPG','HwyMPG','CombMPG']]
sns.pairplot(df2,hue='Cylinders',height=2.5);
```

```python
#scatterplot
sns.scatterplot(x="Eng Displ",y="CombMPG",data=df
            )
```

```python
#joint plot - scatter plot and histogram
sns.jointplot(x="Eng Displ",y="CombMPG",data=df)
```

```python
#lmplot - scatterplot with linear regression
sns.lmplot(x="Eng Displ",y="CombMPG",data=df)
```

```python
#Also has boxplot
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.boxplot(x='Mfr Name', y='CombMPG',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
#swarmplot
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.swarmplot(x='Mfr Name', y='CombMPG',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
#countplot
ax=sns.countplot(x='Mfr Name',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
#heatmap
df2=df.pivot_table(index='Cylinders',columns='Eng Displ',values='CombMPG',aggfunc='mean')
sns.heatmap(df2)
```
