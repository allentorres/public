Train/test
----

Start by creating some data set

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)

x = np.random.normal(3,1,100)
y = np.random.normal(50,30,100) / x
plt.scatter(x,y)
```

Split the data into two

```python
trainX = x[:80]
testX = x[80:]

trainY = y[:80]
testY = y[80:]
```

Here's the training dataset:

```python
plt.scatter(trainX,trainY)
```

Now, let's try to fit an 8th degree polynomial

```python
p8 = np.poly1d(np.polyfit(trainX,trainY,8))
```

Let's plot our polynomial against the training data

```python
xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([-20,200])
plt.scatter(trainX,trainY)
plt.plot(xp,p8(xp),c='r')
plt.show()
from sklearn.metrics import r2_score
print(r2_score(trainY,p8(trainX)))
```

Compare with test data

```python
xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([-20,200])
plt.scatter(testX,testY)
plt.plot(xp,p8(xp),c='r')
plt.show()
from sklearn.metrics import r2_score
print(r2_score(testY,p8(testX)))
```

Activity
---

What is the best degree polynomial for this?

```python
the_plot=[]
for i in range(1,11):
    pn = np.poly1d(np.polyfit(trainX,trainY,i))
    the_plot.append(r2_score(testY,pn(testX)))
plt.plot(range(1,11),the_plot)
```

Seems that the best degree is 6

```python
p6 = np.poly1d(np.polyfit(trainX,trainY,6))
xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([-20,200])
plt.scatter(testX,testY)
plt.plot(xp,p6(xp),c='r')
plt.show()
from sklearn.metrics import r2_score
print(r2_score(testY,p6(testX)))
```
