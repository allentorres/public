Item-based Collaborative Filtering for Movie similarities
===

```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'title','rating']
ratings = pd.read_csv('C:/MLCourse/ml-100k/u-data.txt', sep = '\t', names = r_cols, usecols = range(4))
ratings.head()
```

Now create a pivot table on a dataframe to construct a user/movie rating matrix

```python
movieRatings = ratings.pivot_table(index = ['user_id'],columns = ['title'], values = 'rating')
movieRatings.head()
```

Let's now extract a series of users who rated star wars

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
```

```python
toystoryratings = movieRatings['Toy Story (1995)']
toystoryratings.head()
```

We will use the correlation between the movies with star wars

```python
similarMovies = movieRatings.corrwith(toystoryratings)
df = pd.DataFrame(similarMovies)
df = df.dropna()
#df.head(10)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False).head(20)
```

Correlation is too high for most movie pairings since count is low. Need to give minimum bound

So we can look at stats by movies

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

Let's get rid of movies rated fewer than 100 people to only get the top movies

```python
popularMovies = movieStats['rating']['size'] >= 100
movieStats[popularMovies].sort_values([('rating','mean')],ascending = False)[:15]
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns = ['similarity']))
df.head()
```

```python
df.sort_values(['similarity'], ascending = False)[:15]
```
