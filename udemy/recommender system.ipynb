Item-based Collaborative Filtering for Movie similarities for all users
===

```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'title','rating']
ratings = pd.read_csv('C:/MLCourse/ml-100k/u-data.txt', sep = '\t', names = r_cols, usecols = range(4))
ratings.head()
```

Now create a pivot table on a dataframe to construct a user/movie rating matrix

```python
userRatings = ratings.pivot_table(index = ['user_id'],columns = ['title'], values = 'rating')
userRatings.head()
```

Let's now extract a series of users. Correlation score of every column pair in the matrix

```python
corrMatrix = userRatings.corr(min_periods=100) #should have at least 100 ratings for both movies
corrMatrix.head()
```

Showing all the ratings of user_id = 0

```python
myRatings = userRatings.loc[1].dropna()
myRatings
```

Now, we will be creating recommendations for this user

```python
simCandidates = pd.Series()
for i in range(0,len(myRatings)):
#    print("Adding sims for " + myRatings.index[i] + "...")
    # retrieve similar movies to this one i rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # scale its similarity by how well I rated the movie. just multiply by rating
    sims = sims.map(lambda x: x * myRatings[i])
    # add score to list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
print("sorting...")
simCandidates = simCandidates.groupby(simCandidates.index).sum()
simCandidates.sort_values(inplace = True, ascending = False)
print(myRatings.index)
print(simCandidates)
simCandidates = simCandidates.drop(index=myRatings.index)
simCandidates.head(100)
```
