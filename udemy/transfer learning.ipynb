ResNet50 model for image classification

```python
from IPython.display import Image
```

```python
Image(filename = 'fighterjet.jpg')
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
```

ResNet50 pre-trained CNN expects inputs of 224 x 224. Let's load load up our picture and rescale using the preprocess_input function

```python
img_path = 'fighterjet.jpg'
img = image.load_img(img_path, target_size = (224,224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)
```

```python
model = ResNet50(weights = 'imagenet') #weights indicate which datasets were used to train. This has 1000 classifications
```

Already pre-trained model

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (sample)
print('Predicted:', decode_predictions(preds, top = 3)[0])
```

Let's try another image

```python
def classify(img_path):
    display(Image(filename = img_path))
    
    img = image.load_img(img_path, target_size = (224,224))
    
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis = 0)
    x = preprocess_input(x)
    
    preds = model.predict(x)
    return print('Predicted:', decode_predictions(preds, top = 3)[0])

classify('homer-woohoo.gif')
```
